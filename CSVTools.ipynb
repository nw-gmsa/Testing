```python
import pandas as pd

li=[]
# Read the CSV file
ney_data = pd.read_csv("Input/PDS/NEY.csv")
nw_data = pd.read_csv("Input/PDS/NW.csv")

li.append(ney_data)
li.append(nw_data)

pds = pd.concat(li, axis=0, ignore_index=True)

pds
```

```python
ctdna_data = pd.read_csv("NotGit/StarLIMSSampleData.csv")

combined = pd.concat([ctdna_data, pds], axis=1)
ctdna_data = ctdna_data.iloc[0:0]

combined.dropna(subset=['NHSNumber'], inplace=True)

combined['NHSNumber'] = combined['NHS Number']
combined['PatientFamilyName'] = combined['Surname']
combined['PatientGivenName'] = combined['Forename']
combined.loc[combined['Gender'] == 'M','AdministrativeSex'] = 'Male'
combined.loc[combined['Gender'] == 'F','AdministrativeSex'] = 'Female'
combined['OrderingProviderIdentifier'] = 'Dr_Clinician'
combined['OrderingProviderName'] = 'Dr Clinician'
combined['ClinicalDetails'] = 'REDACTED'
combined['PatientAccessionIdentifier'] = combined['PatientAccessionIdentifier'].astype(int)
#combined['PatientAccessionIdentifier'] = combined['PatientAccessionIdentifier'].map(lambda x: x.rstrip('.0'))

combined['DateOfBirth'] = pd.to_datetime(combined['Date Of Birth'], format='%d/%m/%Y')

combined.drop(columns=['NHS Number','Middle name','Other Name','Deceased', 'GP Surgery', 'Address line 1','Address line 2','Address line 3','Address line 4','Address line 5','Postcode','Surname', 'Forename', 'Gender','Date Of Birth'], inplace=True)
```

```python
combined.to_csv('Input/StarLIMSSampleData.csv', index=False, sep=',')
```
