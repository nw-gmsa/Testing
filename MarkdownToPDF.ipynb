As html

```python
import markdown

html_text = markdown.markdown(md_text)

print(html_text)
```

convert to pdf and save


```python
import base64
from weasyprint import HTML

folder = "Output/PDF/R01/"
filename = "SHIRE_ORU_R01_RM3.txt"


HTML(string=html_text).write_pdf(folder + filename + '.pdf')

byte = HTML(string=html_text).write_pdf()
encoded = base64.b64encode(byte)
encoded = encoded.decode('utf-8')

print(encoded)
```
