```python
import base64
import io

file_list = ['Cancer.Txt']
for file in file_list:
    with open('NotGit/' + file, 'r', encoding='UTF-8') as g:
        while line := g.readline():
            if line.startswith('OBX'):
                obx=line.split('|')
                #print(obx[4])
                if obx[4]=='PDF':
                    ED=obx[5].split('^')
                    #print(ED[2])
                    encoded=ED[4]
                    decode = base64.b64decode(encoded)
                    pdf_stream = io.BytesIO(decode)
                    with open('NotGit/' + file+ '.pdf', 'wb') as pdf:
                        pdf.write(decode)
```
